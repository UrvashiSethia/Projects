Encoding

```python
from sklearn.compose import ColumnTransformer  # to convert into binary form for computer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))           
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
df1.head()
```

```python
#converting numerical columns to standard scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train= sc.fit_transform(X_train[:, 13:]) #for numerical colmns and have to ignore encoded colums of string data
X_test= sc.transform(X_test[:,13:])
```

KNN

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc_list = []
err_list = []
for i in range(1,25):
    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test,y_pred)
    err= 1-acc
    print("For k = {}, accuracy = {}".format(i,acc))
    acc_list.append(acc)
    err_list.append(err)
    #print("For k = {}, error = {}".format(i,err))
```

```python
#Plot line chart for error rate
plt.plot(list(range(1,25)),err_list,c= 'r')
plt.title('Error rate v/s K')
plt.xlabel('K')
plt.ylabel('Error rate')
plt.show()
```

```python
#Model building
model = KNeighborsClassifier(n_neighbors=18)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python

#Model Evaluation
confusion_matrix(y_test,y_pred)
```

```python
#accuracy check
a=accuracy_score(y_test,y_pred)
print(a*100,'%')
```
